### 2.0. 패키지 불러오기 & 데이터 프레임 불러오기


```python

### 0. 패키지 불러오기 & 데이터 프레임 불러오기


#패키지 불러오기
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import  linear_model
from sklearn import  preprocessing
from pandas.plotting import scatter_matrix
```

```python
# Train - dataset 데이터 프레임 정의
data_url ='train.csv'
df_train=pd.read_csv(data_url, header = 'infer')

df_train.head()
```

```python
# Test - dataset 데이터 프레임 정의
data_url ='test.csv'
df_test=pd.read_csv(data_url, header = 'infer')

df_test.head()
```

```python
# Y 값 (목적값) 을 따로 빼서 정리 del 보다 번거롭지 않아서 pop이 좋은듯.
y_data = df_train.pop("SalePrice")
y_data.head()
```

```python
# .shape 명령어를 사용해서 각 데이터셋의 행렬 사이즈를 체크함.
df_train.shape, df_test.shape
```


### 2.1.1 결측치 처리


```python
# 두 데이터 셋을 합쳐주기 (id도 연달아서 매겨져 있기 떄문에 합쳐주는게 타당한 듯)
df_all = pd.concat([df_train, df_test])

df_all.head()
```

```python
# 결측치가 존재하는 행 순서대로 나타내기.

df_all.isnull().sum().sort_values(ascending=False).head(34)
```

```python
# 결측치 높은 순서 부터 비율을 DataFrame으로 나타냄

missing_value_rank = df_all.isnull().sum() / df_all.isnull().count()
missing_value_rank_df = pd.DataFrame(missing_value_rank, columns=['Missing Value Ratio'])
missing_value_rank_df[missing_value_rank_df['Missing Value Ratio'] > 0].sort_values('Missing Value Ratio', ascending=False)
```

```python
# 원래는 각 데이터마다 특성을 파악하고 컬럼마다 각기 판단해야한다.
# 하지만 결측치가 데이터의 대부분인 경우 나머지 자료를 유추하는 건 타당하지 않다고 판단해서 상위 5개 컬럼은 삭제하기로 결정함.
# 결측퍼센트가 40퍼센트 이상인 놈들은 전부 소거

for column_name in df_all.columns[df_all.isnull().sum() / df_all.isnull().count() > 0.40]:
    del df_all[column_name]
df_all.head()
```

```python
## 오브젝트형태의 데이터와 아닌 데이터를 분류하여 결측치 별개 처리 // 각 평균, 최빈값은 각기 missing_values 딕셔너리에 저장하여 보관

missing_values = {} 
for column_name in df_all.columns:
    if df_all[column_name].dtypes == "object":
        mode_value = df_all[column_name].value_counts()[0]    #  오브젝트 데이터이기 때문에 해당 컬럼의 최빈값으로 편성
        missing_values[column_name] = mode_value
        df_all[column_name] = df_all[column_name].fillna(mode_value)
    else:
        mean_value = df_all[column_name].mean()  #수치기 떄문에 해당 컬럼의 평균값으로 대체
        missing_values[column_name] = mean_value
        df_all[column_name] = df_all[column_name].fillna(mean_value)
```

```python
# 결측치 처리 완료

df_all.isnull().sum().sum()
```


### 2.1.2. 이산형 데이터 처리


```python
#이산형 데이터를 처리한다.

df_new_all = pd.get_dummies(df_all)
```


### 2.1.3. 데이터 스케일링


```python
min_max_values = {a:[b,c,d] for a, b, c, d in zip(list(df_new_all.columns[1:]), df_new_all.max()[1:], df_new_all.min()[1:], df_new_all[1:].mean())}
min_max_values
```

```python

# 최대최소 스케일러를 통해서 데이터 스케일링 모듈로도 가능

for column_name in min_max_values.keys():
    df_new_all[column_name] =  ( df_new_all[column_name] - min_max_values[column_name][1] ) \
        / (min_max_values[column_name][0] - min_max_values[column_name][1]) * (5 - 1) + 1
```


### 2.2. 모델에 수치 대입 후 파라미터 및 R 값 도출


```python

# id 컬럼을 제외한 1459번 까지의 row를 train 값으로 편성
x_train = df_new_all.iloc[:1460,1:].as_matrix()
# id 컬럼을 제외한 1459번 부터의 row를 test 값으로 편성
x_test = df_new_all.iloc[1460:,1:].as_matrix()
y_train = y_data.as_matrix()
```

```python
from sklearn import linear_model

#선형회귀식에 x, y train 값을 주어 학습
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mae = mean_absolute_error(y_train, regr.predict(x_train))
mse = mean_squared_error(y_train, regr.predict(x_train))
r2 = r2_score(y_train, regr.predict(x_train))

print(mae, mse, r2)
```
